```python
#시도 이름 바꾸기

def SIDOname_change(SIDO):
    if((SIDO.find('경기')>-1)):
        SIDO='경기'
    elif ((SIDO.find('인천')>-1)):
        SIDO='인천'
    elif ((SIDO.find('서울')>-1)):
        SIDO='서울'
    else:
        SIDO
    
    return SIDO

#명칭이 바뀐 시군구 이름 바꾸기

def SGGname_change(SIDO, SGG):
    if((SIDO.find('경기')>-1)&(SGG=='여주군')):
        SGG='여주시'
    elif ((SIDO.find('인천')>-1)&(SGG=='남구')):
        SGG='미추홀구'
    else:
        SGG
    
    return SGG
```

```python
# 오류 행정동 코드 바꾸기
def SGGcode_change(SGG_CD):
    if(SGG_CD=='23030'):
        SGG_CD='23090'
    else:
        SGG_CD
    
    return SGG_CD
```

```python
existing_data['행정구역코드'] = existing_data['행정구역코드'].apply(lambda x:SGGcode_change(x))

existing_data
```

```python
def change_BJcity_code(BJcity):
    BJcity=str(BJcity)
    
    if(BJcity[0:2]=='41'):
        BJcity = BJcity[0:4]+"0" # 경기도 구단위를 전부 시단위로 올려주기
    else:
        BJcity
    
    return BJcity
```

```python
busstop_df = pd.DataFrame()

busstop_df1 = pd.read_csv('../data/traffic/bus_route1.csv').drop(columns=['citycode'])
busstop_df2 = pd.read_csv('../data/traffic/bus_route2.csv')
busstop_df3 = pd.read_csv('../data/traffic/bus_route3.csv').drop(columns=['citycode'])

busstop_df = busstop_df.append(busstop_df1)
busstop_df = busstop_df.append(busstop_df2)
busstop_df = busstop_df.append(busstop_df3)

busstop_df = busstop_df.drop_duplicates()
busstop_df = busstop_df.rename(columns={"A1":"법정구역코드"})
busstop_df['법정구역코드'] = busstop_df['법정구역코드'].apply(lambda x:change_BJcity_code(x))

busstop_count = busstop_df[['법정구역코드','routeid','nodeid']].groupby(by='법정구역코드').nunique().reset_index().rename(columns={"routeid":"버스_노선_개수","nodeid":"버스_정류장_개수"}) #고유 값 갯수 세기
busstop_count
```

```python
concat_data = pd.merge(concat_data, busstop_count, on ='법정구역코드',how='left')
concat_data
```
